## 1. List Operations and Basic Statistics:<br/>
* Create a list of the first 20 natural numbers.<br/>
* Compute the sum of all even numbers in the list.<br/>
* Compute the product of all odd numbers in the list.<br/>
* Calculate the mean, median, and standard deviation of the list.<br/>


```python
list_first_20 = list(range(1,21))
print(f"list_first_20 = {list_first_20}")


even_sum = 0
odd_sum = 0
for ele in list_first_20:
    if ele%2 == 0:
        even_sum += ele
    else:
        odd_sum += ele
print(f"Sum of even numbers : {even_sum}")
print(f"Sum of odd numbers : {odd_sum}")

n = len(list_first_20)
mean = sum(list_first_20)/n
median = (list_first_20[n//2-1]+list_first_20[n//2])/2
variance = sum((x - mean) ** 2 for x in list_first_20) / (n - 1)
std_dev = variance ** 0.5

print(f"Mean : {mean}")
print(f"Median : {median}")
print(f"Standard deviation : {std_dev}")
```

## 6. Palindrome Check:
* Write a program to check if a string is a palindrome (ignoring spaces and case).<br/>
* Input: "A man a plan a canal Panama" <br/>
* Output: True <br/>

```python
input_string = input("Enter a string :")
l = 0
r = len(input_string)-1
ans = True
while l <= r :
    if input_string[l] == ' ':
        l +=1
    if input_string[r] == ' ':
        r -=1
    if input_string[l].lower() != input_string[r].lower():
        ans = False
        break
    else:
        l +=1
        r -=1


print(f'Output : {ans}')
    
```

## 9 . Merging Dictionaries:
* Write Python code to merge two dictionaries. If a key is present in both, sum their values.<br/>
<br/>
* Example:<br/>
dict1 = {'a': 10, 'b': 20}<br/>
dict2 = {'b': 5, 'c': 15}<br/>
Output: {'a': 10, 'b': 25, 'c': 15}<br/>

```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}

merge_dict  = dict1.copy()

for key,value in dict2.items():
    if key in merge_dict:
        merge_dict[key] += value
    else:
        merge_dict[key] = value

print(f'Merged Dictnory : {merge_dict}')
```

## 10 . Character Frequency:
Write a function that accepts a string and:
* Converts all characters to lowercase.<br/>
* Counts the frequency of each character.<br/>
* Returns a dictionary with characters as keys and their counts as values.<br/>
* Visualize the character frequencies using a bar chart.<br/>

```python

def freq_char(input_str):
    input_str = input_str.lower()

    freq_dict = {}
    for char in input_str:
        if char in freq_dict:
            freq_dict[char]+=1
        else:
            freq_dict[char] = 1

    return freq_dict

input_str = input("Enter a string")
freq_dict = freq_char(input_str)

print(freq_dict)

characters = list(freq_dict.keys())
frequencies = list(freq_dict.values())

plt.bar(characters, frequencies)
plt.title('Character Frequency Distribution')
plt.xlabel('Character')
plt.ylabel('Frequency')
plt.show()
```

## 11. Remove Duplicates:
* Write a program to remove duplicate elements from a list while maintaining the order of elements.<br/>
* Calculate the mean and median of the unique elements.

```python
def remove_duplicates(input_list):
    unique_list = []
    
    # Traverse through the input list and add elements to the unique list if not already present
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    
    return unique_list

def calculate_mean_median(unique_list):
    mean = sum(unique_list) / len(unique_list) if unique_list else 0
    median = np.median(unique_list) if unique_list else 0
    return mean, median

# Example list with duplicates
input_list = [10, 20, 10, 30, 40, 20, 30, 50]

# Remove duplicates while maintaining order
unique_list = remove_duplicates(input_list)

# Calculate mean and median of unique elements
mean, median = calculate_mean_median(unique_list)

# Output results
print(f"Unique List: {unique_list}")
print(f"Mean: {mean}")
print(f"Median: {median}")
```

## 12. String Operations:
Given a list of strings, write a Python function to: <br/>
*   Find the longest string.<br/>
*   Count how many strings start with a vowel.<br/>
*   Visualize the distribution of string lengths using a histogram.<br/>


```python
string_list = ["apple","ball","cat","dog","elephant","ink"]

longest_string = max(string_list, key=len)
    

vowel_list = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']

count_starting_with_vowel = sum(1 for s in string_list if s[0] in vowel_list)

lengths = [len(s) for s in string_list]

print(f"Longest string: {longest_string}")
print(f"Number of strings starting with a vowel: {count_starting_with_vowel}")


plt.hist(lengths, bins=range(min(lengths), max(lengths) + 1))
plt.title('Distribution of String Lengths')
plt.xlabel('Length of String')
plt.ylabel('Frequency')
plt.show()
```

## 15. Anagram Finder:
* Write a Python function to find all anagrams of a word in a given list.<br/>
* Input: word = "listen", words = ["enlist", "google", "inlets", "banana"]<br/>
* Output: ["enlist", "inlets"]<br/>

```python
def find_anagrams(word, words_list):
    
    sorted_word = sorted(word)
    return [w for w in words_list if sorted(w) == sorted_word]


word = "listen"
words = ["enlist", "google", "inlets", "banana"]


anagrams = find_anagrams(word, words)

print(anagrams)
```

## 17. File Operations:
Write Python code to:<br/>
* Create a text file.<br/>
* Write a list of numbers to the file (one per line).<br/>
* Read the file and compute their sum.

```python
def write_and_sum_numbers(filename, numbers):
    with open(filename, 'w') as file:
        for number in numbers:
            file.write(f"{number}\n")

    with open(filename, 'r') as file:
        numbers_from_file = [int(line.strip()) for line in file.readlines()]
    
    total_sum = sum(numbers_from_file)
    
    return total_sum


numbers = [10, 20, 30, 40, 50]
filename = 'numbers.txt'


result = write_and_sum_numbers(filename, numbers)
print(f"The sum of the numbers is: {result}")
```

## 19. Random Number Generation:
Write a Python program to:
* Generate 100 random integers between 1 and 50.
* Save them in a text file, one per line.<br/>
* Read the file and compute the mean of the numbers.<br/>
* Visualize the distribution of the numbers using a histogram.<br/>

```python
import random

def generate_and_analyze_numbers(filename):
    
    random_numbers = [random.randint(1, 50) for _ in range(100)]

    with open(filename, 'w') as file:
        for number in random_numbers:
            file.write(f"{number}\n")
    
    with open(filename, 'r') as file:
        numbers_from_file = [int(line.strip()) for line in file.readlines()]
    
    mean = sum(numbers_from_file) / len(numbers_from_file)
 
    print(f"The mean of the numbers is: {mean}")

    plt.hist(numbers_from_file, bins=range(1, 52))
    plt.title('Distribution of Random Numbers')
    plt.xlabel('Number')
    plt.ylabel('Frequency')
    plt.show()

filename = 'random_numbers.txt'

generate_and_analyze_numbers(filename)
```

```python
import numpy as np
import matplotlib.pyplot as plt


numbers = [12, 15, 14, 10, 13, 10, 15, 14, 16, 12, 14, 13, 18, 20, 12]


mean = np.mean(numbers)
median = np.median(numbers)
values, counts = np.unique(numbers, return_counts=True)
mode = values[counts.argmax()]
std_dev = np.std(numbers)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")
```

## 22. Normalization:
* Write a Python function to normalize a list of numbers to a range of 0 to 1.<br/>
* Visualize the normalized data using a line plot.

```python
numbers = [12, 15, 14, 10, 13, 10, 15, 14, 16, 12, 14, 13, 18, 20, 12]

min_value = np.min(numbers)
max_value = np.max(numbers)
normalized_numbers = [(x - min_value) / (max_value - min_value) for x in numbers]

print("Normalized Data:", normalized_numbers)

plt.plot(normalized_numbers)
plt.title('Normalized Data (0 to 1)')
plt.xlabel('Index')
plt.ylabel('Normalized Value')
plt.show()
```

* Write a Python program that takes a year as input and checks if it is a leap year.

```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```
